# Analise exploratoria - Ecommerce Estatistica

Este notebook le o arquivo `ecommerce_estatistica.csv`, faz uma analise inicial e gera os 7 graficos solicitados:

1. Histograma  
2. Dispersao  
3. Mapa de calor  
4. Barra  
5. Pizza  
6. Densidade  
7. Regressao  

Tambem inclui comentarios curtos com os principais insights para ajudar na entrega.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```
# Leitura dos dados
df = pd.read_csv('/mnt/data/ecommerce_estatistica.csv')

# Remocao de coluna indice, se existir
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Visualizacao inicial
print('Dimensoes:', df.shape)
display(df.head())
```

```
# Tipos de dados e estatisticas iniciais
display(df.info())
display(df.describe(include='all').transpose())
```

## Destaques identificados na base

Durante a exploracao dos dados, alguns pontos chamam a atencao:

- `N_Avaliacoes` e `Qtd_Vendidos_Cod` tem grande dispersao e assimetria.
- A correlacao mais forte com quantidade vendida tende a aparecer em `N_Avaliacoes`.
- `Preco`, `Desconto` e `Nota` parecem ter relacoes mais fracas com volume vendido.
- As categorias de `Marca`, `Material` e `Temporada` permitem bons recortes de distribuicao.

A seguir, os graficos solicitados.


```
# 1) Grafico de Histograma - distribuicao do preco
plt.figure(figsize=(10, 6))
sns.histplot(df['Preço'], bins=20, kde=False)
plt.title('Histograma do Preco dos Produtos')
plt.xlabel('Preco')
plt.ylabel('Frequencia')
plt.show()
```

```
# 2) Grafico de Dispersao - preco x quantidade vendida
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Preço', y='Qtd_Vendidos_Cod')
plt.title('Grafico de Dispersao: Preco x Quantidade Vendida')
plt.xlabel('Preco')
plt.ylabel('Quantidade Vendida (codigo)')
plt.show()
```

```
# 3) Mapa de Calor - correlacao entre variaveis numericas
plt.figure(figsize=(12, 8))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor das Correlacoes')
plt.xlabel('Variaveis')
plt.ylabel('Variaveis')
plt.show()
```

```
# 4) Grafico de Barra - top 10 marcas por frequencia
top_marcas = df['Marca'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_marcas.index, y=top_marcas.values)
plt.title('Top 10 Marcas por Quantidade de Produtos')
plt.xlabel('Marca')
plt.ylabel('Quantidade de Produtos')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```
# 5) Grafico de Pizza - distribuicao por temporada
temporada_counts = df['Temporada'].value_counts().head(6)

plt.figure(figsize=(8, 8))
plt.pie(
    temporada_counts.values,
    labels=temporada_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Distribuicao dos Produtos por Temporada')
plt.show()
```

```
# 6) Grafico de Densidade - distribuicao do desconto
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Desconto', fill=True)
plt.title('Grafico de Densidade do Desconto')
plt.xlabel('Desconto')
plt.ylabel('Densidade')
plt.show()
```

```
# 7) Grafico de Regressao - numero de avaliacoes x quantidade vendida
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='N_Avaliações', y='Qtd_Vendidos_Cod', scatter_kws={'alpha': 0.6})
plt.title('Grafico de Regressao: Numero de Avaliacoes x Quantidade Vendida')
plt.xlabel('Numero de Avaliacoes')
plt.ylabel('Quantidade Vendida (codigo)')
plt.show()
```

```
# Correlacoes com quantidade vendida para enriquecer a analise
corr_vendas = (
    df.select_dtypes(include=np.number)
      .corr()['Qtd_Vendidos_Cod']
      .sort_values(ascending=False)
)

display(corr_vendas.to_frame('correlacao_com_qtd_vendida'))
```

## Conclusoes sugeridas para a entrega

Com base nos graficos e nas correlacoes, voce pode destacar:

- A quantidade de avaliacoes possui forte relacao positiva com a quantidade vendida.
- O preco nao parece explicar sozinho o volume de vendas.
- O desconto apresenta distribuicao concentrada em faixas especificas.
- Algumas marcas aparecem com maior concentracao de produtos na base.
- A regressao reforca que produtos com mais avaliacoes tendem a estar associados a maiores volumes de venda.

Esses achados ajudam a direcionar decisoes de negocio sobre sortimento, prova social, politica de desconto e destaque de produtos.
